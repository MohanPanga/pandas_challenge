```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas DataFrames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset.  
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
school_data_complete.head()
```

## District Summary

* Calculate the total number of schools

* Calculate the total number of students

* Calculate the total budget

* Calculate the average math score 

* Calculate the average reading score

* Calculate the percentage of students with a passing math score (70 or greater)

* Calculate the percentage of students with a passing reading score (70 or greater)

* Calculate the percentage of students who passed math **and** reading (% Overall Passing)

* Create a dataframe to hold the above results

* Optional: give the displayed data cleaner formatting

```python
# Distric total schools, students and budget
Total_Schools = school_data['school_name'].nunique()
Total_Students = student_data['student_name'].count()
Total_Budget = school_data['budget'].sum()

# District average scores
Avg_math_score = student_data['math_score'].mean()
Avg_math_score = "{:.2f}".format(Avg_math_score)
Avg_reading_score = student_data['reading_score'].mean()

# Perccent passed in math, reading and overall
passed_math = student_data[student_data['math_score']>=70]['student_name'].count()
passed_math_percent = passed_math/Total_Students*100

passed_reading = student_data[student_data['reading_score']>=70]['student_name'].count()
passed_reading_percent = passed_reading/Total_Students*100

passed_overall = student_data[(student_data['math_score']>=70) & (student_data['reading_score']>=70)]['student_name'].count()
passed_overall_percent = passed_overall/Total_Students*100

# formatting data before loading to dataframe
format_data = [Avg_reading_score, passed_math_percent, passed_reading_percent, passed_overall_percent ]
[Avg_reading_score, passed_math_percent, passed_reading_percent, passed_overall_percent ] = list(map("{:.2f}".format, format_data))
Total_Students ="{:,}".format(Total_Students)
Total_Budget = "${:,.2f}".format(Total_Budget)

# creating dataframe for the results
result_df = pd.DataFrame({'Total_Schools':[Total_Schools],
                         'Total_Students':Total_Students,
                          'Total Budget':Total_Budget,
                         'Average Math Score':Avg_math_score,
                         'Average Reading Score':Avg_reading_score,
                         'Percent Passing Math':passed_math_percent,
                         'Percent Passing Reading':passed_reading_percent,
                         'Percent Overall Passing':passed_overall_percent})
result_df
```

## School Summary

* Create an overview table that summarizes key metrics about each school, including:
  * School Name
  * School Type
  * Total Students
  * Total School Budget
  * Per Student Budget
  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * % Overall Passing (The percentage of students that passed math **and** reading.)
  
* Create a dataframe to hold the above results

```python
# School name, type, total students and school budget can be obtained from school_data dataframe

# creating a new dataframe summary_df from school_data
col1 =['school_name','type','size','budget']
summary_df = school_data[col1].copy(deep=True)

# calculating per student budget
summary_df['Per Student Budget'] = summary_df['budget']/summary_df['size']
summary_df.set_index('school_name',inplace =True)

# to calculate average math and reading scores per school, student_data is grouped by school name and mean function is applied to the scores
group_student = student_data.groupby('school_name')
summary_df[['Average Math_score','Average Reading Score']] = group_student[['math_score','reading_score']].mean()

# to calculate % passing math, reading and overall for each school, student data is filtered by passing criteria and then grouped by school name.
summary_df['Students Passing Math'] = student_data[student_data['math_score']>=70].groupby('school_name')['student_name'].count()
summary_df[['% Passing Math']] = summary_df['Students Passing Math']/summary_df['size']*100

summary_df['Students Passing Reading'] = student_data[student_data['reading_score']>=70].groupby('school_name')['student_name'].count()
summary_df['% Passing Reading'] = summary_df['Students Passing Reading']/summary_df['size']*100

summary_df['Students Passing Overall'] = student_data[(student_data['reading_score']>=70) & (student_data['math_score']>=70)].groupby('school_name')['student_name'].count()
summary_df['% Passing Overall'] = summary_df['Students Passing Overall']/summary_df['size']*100

# format results
summary_df['budget'] = summary_df['budget'].map("${:,}".format)
summary_df['Per Student Budget'] = summary_df['Per Student Budget'].map("${:,}".format)

# School summary
summary_df[['type','size','budget','Per Student Budget','Average Math_score', 'Average Reading Score','% Passing Math','% Passing Reading', '% Passing Overall']]
```

## Top Performing Schools (By % Overall Passing)

* Sort and display the top five performing schools by % overall passing.

```python
Top_schools = summary_df.sort_values('% Passing Overall', ascending = False)
Top_schools.head(5)
```

## Bottom Performing Schools (By % Overall Passing)

```python
Bottom_schools = summary_df.sort_values('% Passing Overall', ascending = True)
Bottom_schools.head(5)
```

## Math Scores by Grade

* Create a table that lists the average Reading Score for students of each grade level (9th, 10th, 11th, 12th) at each school.

  * Create a pandas series for each grade. Hint: use a conditional statement.
  
  * Group each series by school
  
  * Combine the series into a dataframe
  
  * Optional: give the displayed data cleaner formatting

```python
# create list of grades for which average score is needed
grades = ['9th','10th','11th','12th']

# create empty data frame to collect the results
grade_mathscore_df = pd.DataFrame()

#loop over the list of grades and store the result in grade_mathscore_df
for x in grades:
    grade_mathscore_df[x] = student_data[student_data['grade'] == x].groupby('school_name')['math_score'].mean()

grade_mathscore_df
```

## Reading Score by Grade 

* Perform the same operations as above for reading scores

```python
# create empty data frame to collect the results
grade_readingscore_df = pd.DataFrame()

#loop over the list of grades and store the result in grade_readingscore_df
for x in grades:
    grade_readingscore_df[x] = student_data[student_data['grade'] == x].groupby('school_name')['reading_score'].mean()

grade_readingscore_df
```

## Scores by School Spending

* Create a table that breaks down school performances based on average Spending Ranges (Per Student). Use 4 reasonable bins to group school spending. Include in the table each of the following:
  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * Overall Passing Rate (Average of the above two)

```python
# make a copy of summary_df dataframe that has school summary
score_by_school_spend = summary_df.copy(deep = True)

# cast budget column to folat
score_by_school_spend['Per Student Budget'] = score_by_school_spend['Per Student Budget'].str.replace("$","").astype(float)

# create bins for spending per student and corresponding labels
bins =[0, 584, 629, 644, 675]
labels =['<584','585-629','630-644','644-675']

# bin data on school spending per student and group by spending range
score_by_school_spend['Spending Range'] = pd.cut(score_by_school_spend['Per Student Budget'], bins, labels=labels,include_lowest=True)
group_score_by_school_spend = score_by_school_spend.groupby('Spending Range')[['Average Math_score', 'Average Reading Score', '% Passing Math','% Passing Reading','% Passing Overall']].mean()

# round data for dispaly
group_score_by_school_spend.round(2)
```

## Scores by School Size

* Perform the same operations as above, based on school size.

```python
# make a copy of summary_df that contains summary of school data
score_school_size = summary_df.copy(deep = True)

# create bins based on school size and create labels for the bins
bin_size = [0, 1000, 2000, 5000]
label_size =['Small(<1000)', 'Medium (1000-2000)','Large (2000-5000)']

# bin the data based on size into School Size column
score_school_size['School Size'] = pd.cut(score_school_size['size'], bin_size, labels = label_size, include_lowest=True)
group_school_size = score_school_size.groupby('School Size')[['Average Math_score', 'Average Reading Score', '% Passing Math','% Passing Reading','% Passing Overall']].mean()

# format the result
group_school_size.round(2)
```
